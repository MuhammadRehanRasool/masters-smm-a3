# 1. Clean Data and Prepare Final Dataset

```python
import pandas as pd
import os
from datetime import datetime
import numpy as np
 
# Define file paths
data_dir = 'raw_dataset'
files = ['islamabad_complete_data.xlsx', 'karachi_complete_data.xlsx', 'lahore_complete_data.xlsx']
cities = ['Islamabad', 'Karachi', 'Lahore']
 
dfs = []
for file, city in zip(files, cities):
    df = pd.read_excel(os.path.join(data_dir, file))
    # Extract date from datetime
    df['date'] = pd.to_datetime(df['datetime']).dt.date
    # Add city column
    df['city'] = city
    # Rename columns
    df = df.rename(columns={
        'components.pm2_5': 'pm2_5',
        'components.pm10': 'pm10',
        'components.no2': 'no2',
        'temperature_2m': 'temp',
        'relative_humidity_2m': 'humidity'
    })
    # Select relevant columns
    df = df[['date', 'city', 'pm2_5', 'pm10', 'no2', 'temp', 'humidity']]
    dfs.append(df)
 
# Combine all dataframes
combined_df = pd.concat(dfs, ignore_index=True)
 
# Fill missing values with mean of respective columns
for col in ['pm2_5', 'pm10', 'no2', 'temp', 'humidity']:
    combined_df[col] = combined_df[col].fillna(combined_df[col].mean())
    

combined_df['date'] = pd.to_datetime(combined_df['date'])
```
